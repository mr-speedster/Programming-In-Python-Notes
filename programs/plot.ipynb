```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [50,60,70,80,90]
plt.plot(x,y,color='red',linewidth=5,linestyle='dashed')
plt.xlabel('Distance')
plt.ylabel('Time')
plt.title('Speed')
```

```python
plt.plot(x,y,'g*')
```

```python
plt.plot(x,y,'--g*')
```

```python
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [25,25,35,45,55]
y3 = [10,30,50,70,90]
plt.plot(x,y1,color='red',label="T1")
plt.plot(x,y2,color='blue',label="T2")
plt.plot(x,y3,color='green',label="T3")
plt.xlabel('Time')
plt.ylabel('Distance')
plt.legend(loc='best')
```

### Pi chart

```python
import numpy as np
```

```python
values = np.array([10,50,20,15,5])
label = ["A","B",'C','D','E']
plt.pie(values,labels=label)
plt.show()
```

```python
company=['Google','Amazon','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,80,34,12]
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,label='Revenue',width=0.4)
plt.bar(xpos+0.2,profit,label='Profit',color='red',width=0.4)
plt.legend(loc='best')
```
